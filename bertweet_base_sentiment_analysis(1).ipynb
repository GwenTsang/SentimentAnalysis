```python
import os
from huggingface_hub import InferenceClient

client = InferenceClient(
    provider="auto",
    api_key=os.environ["HF_TOKEN"],
)

result = client.text_classification(
    "Last week, I saw my psychologist, and I must say that I was able to smile again. Yes, a smile, even joy. Well, come on, what matters, after all, is happiness. But right now I'm not doing well: I want to kill myself, slit my throat with a knife, torture people to death. ",
    model="finiteautomata/bertweet-base-sentiment-analysis",
)

print(result)
```

**Hypothesis**: the end of the tweet seems to carry slightly more weight in the decision. Indeed, the end of a tweet or short text often signals the conclusion, which is often one of the most important elements. Admittedly, we can see that the result of the last test is ambivalent and balanced.
